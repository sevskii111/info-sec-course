```python
import matplotlib.pyplot as plt
```

```python
class Linear_congruentiol_generator:
    def __init__(self, a, c, seed = 1, m = 32, n = 32):
        self.__m = m
        self.__seed = seed
        self.__a = a
        self.__c = c
        self.__n = n
        self.__counter = 0

    def __iter__(self):
        return self
    
    def __next__(self):
        rand = (self.__a * self.__seed + self.__c) % self.__m
        self.__seed = rand
        self.__counter += 1
        if self.__counter > self.__len__():
            raise StopIteration
        return rand

    def __len__(self):
        return self.__n
```

```python
nums = []
for i in Linear_congruentiol_generator(17, 5, n=10000):
    #print(i)
    nums.append(i)
```

```python
plt.hist(nums, bins=32)
```
